```python
import matplotlib
import numpy as np
import pandas as pd
from matplotlib import gridspec
import matplotlib.pyplot as plt

%matplotlib inline  

matplotlib.rcParams.update({'font.size': 12})
```

# Lambeck et al. - sea-level

```python
filename = 'data/lambeck.txt'
sldata = pd.read_csv(filename, sep=r'\s+', engine='c', na_filter=False, \
                                   dtype=np.float, low_memory=False)
```

```python
age = -sldata.values[:,0]*1000.
lower = sldata.values[:,1]
upper = sldata.values[:,2]
ave = lower+(upper-lower)/2.
```

```python
size = (10,7)
dpi = 80
font = 10

# Define figure size
fig, ax = plt.subplots(1,figsize=size, dpi=dpi)
ax.set_facecolor('#f2f2f3')

# Plotting curves
ax.plot(age, upper, label='upper limit',linewidth=2,c='gray',ls=':') #,c='#415cf4')
ax.plot(age, lower, label='lower limit',linewidth=2,c='gray',ls='--')#'#42b0f4')
ax.plot(age, ave, label='average',linewidth=2,c='#415cf4')

ax.fill_between(age, lower, upper, facecolor='lightgray')

# Legend, title and labels
plt.grid()
lgd = plt.legend(frameon=False,loc=4,prop={'size':font+1}, bbox_to_anchor=(1.2,0.2))
plt.xlabel('Time [y]',size=font+2)
plt.ylabel('relative sea-level [m]',size=font+2)
plt.xlim(-140000, 0.)


ttl = ax.title
ttl.set_position([.5, 1.0])
plt.title('Lambeck sea-level curve',size=font+3)
plt.show()
```

# Grant et al. - sea-level

```python
filename = 'data/grantetal.txt'
sldata = pd.read_csv(filename, sep=r'\s+', engine='c', na_filter=False, \
                                   dtype=np.float, low_memory=False)
```

```python
age = np.flip(-sldata.values[:,0]*1000.,0)
sl = np.flip(sldata.values[:,1],0)
```

```python
size = (12,7)
dpi = 80
font = 10

# Define figure size
fig, ax = plt.subplots(1,figsize=size, dpi=dpi)
ax.set_facecolor('#f2f2f3')

# Plotting curve
ax.plot(age, sl, linewidth=2,c='#415cf4')

# Legend, title and labels
plt.grid()
#lgd = plt.legend(frameon=False,loc=4,prop={'size':font+1}, bbox_to_anchor=(1.2,0.2))
plt.xlabel('Time [y]',size=font+2)
plt.ylabel('relative sea-level [m]',size=font+2)
plt.xlim(-140000, 0.)


ttl = ax.title
ttl.set_position([.5, 1.0])
plt.title('Grant sea-level curve',size=font+3)
plt.show()
```

```python
df = pd.DataFrame({'time':age.flatten(),'sl':sl.flatten()})
df.to_csv('grantetal.csv',columns=['time','sl'], sep=' ', index=False ,header=0)
```
